```python
# AND 게이트 구현하기
# 입력 x1과 x2가 모두 참일때 1을 출력한다

def AND(x1,x2):
  w1,w2,theta = 1,1,1.9
  value = w1*x1 + w2*x2
  if value <= theta:
    return 0 
  elif value > theta:
    return 1
```

```python
print(AND(1,1))
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))

#1,0,0,0
```

```python
def HAND(x1,x2):
  w1,w2,theta = 1,1,1.9
  value = w1*x1 + w2*x2
  if value <= theta:
    return 1
  elif value > theta:
    return 0
```

```python
print(HAND(1,1))
print(HAND(0,0))
print(HAND(1,0))
print(HAND(0,1))
```

```python
def OR(x1,x2):
  w1,w2,theta = 1,1,0
  value = w1*x1 + w2*x2
  if value <= theta:
    return 0
  elif value > theta:
    return 1
    
```

```python
print(OR(1,1))
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
```
